## Pembuatan Model

```python
import pandas as pd

df = pd.read_csv("/content/data_career_changed.csv")
```

```python
df
```

```python
df['Career'].value_counts()
```

```python
df_copy = df.copy()

test_index = [45, 73, 42, 49]
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['Career'] = encoder.fit_transform(df['Career'])

test_df = df.iloc[test_index]
train_df = df.drop(test_index)

y_train = train_df.pop('Career')
X_train = train_df

y_test = test_df.pop('Career')
X_test = test_df
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
      tf.keras.layers.Dense(512, activation = 'relu'),
      tf.keras.layers.Dense(256, activation = 'relu'),
      tf.keras.layers.Dense(128, activation = 'relu'),
      tf.keras.layers.Dense(76, activation = 'softmax')    # < softmax activation here
  ])
model.compile(
  loss=tf.keras.losses.SparseCategoricalCrossentropy(),
  optimizer=tf.keras.optimizers.Adam(0.001),
)

model.fit(
  X_train,y_train,
  epochs=100,
  verbose = 1
)
```

```python
tf.saved_model.save(model, '/content/saved-models')
```
